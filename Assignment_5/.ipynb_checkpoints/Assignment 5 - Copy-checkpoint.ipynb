#### Normalize the pixel values of the images to a range of [0, 1]

```python
x_train = x_train.astype("float32") / 255.0
x_test = x_test.astype( "float32") / 255.0
x_train = x_train[:, None, :, :]
x_test = x_test[:, None, :, :]
```

```python
x_train_tensor = torch.tensor(x_train, dtype=torch.float32)
y_train_tensor = torch.tensor(y_train, dtype=torch.long)
x_test_tensor = torch.tensor(x_test, dtype=torch.float32)
y_test_tensor = torch.tensor(y_test, dtype=torch.long)

mean = x_train_tensor.mean()
std = x_train_tensor.std()
mean, std
```

```python
x_train_tensor = (x_train_tensor - mean) / std
x_test_tensor = (x_test_tensor - mean) / std
```

```python
full_train_dataset = TensorDataset(x_train_tensor, y_train_tensor)

# Split into train + val
train_size = int(0.9 * len(full_train_dataset))
val_size = len(full_train_dataset) - train_size
train_dataset, val_dataset = random_split(full_train_dataset, [train_size, val_size])
test_dataset = TensorDataset(x_test_tensor, y_test_tensor)

BATCH_SIZE = 32
train_dataloader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
val_dataloader = DataLoader(val_dataset, batch_size=BATCH_SIZE, shuffle=False)
test_dataloader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

len(train_dataloader), len(val_dataloader), len(test_dataloader) 
```

```python
class LeNet5V1(nn.Module):
    def __init__(self):
        super().__init__()
        self.feature = nn.Sequential(
            #1
            nn.Conv2d(in_channels=1, out_channels=6, kernel_size=5, stride=1, padding=2),   # 28*28->32*32-->28*28
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),  # 14*14
            
            #2
            nn.Conv2d(in_channels=6, out_channels=16, kernel_size=5, stride=1),  # 10*10
            nn.Tanh(),
            nn.AvgPool2d(kernel_size=2, stride=2),  # 5*5
            
        )
        self.classifier = nn.Sequential(
            nn.Flatten(),
            nn.Linear(in_features=16*5*5, out_features=120),
            nn.Tanh(),
            nn.Linear(in_features=120, out_features=84),
            nn.Tanh(),
            nn.Linear(in_features=84, out_features=10),
        )
        
    def forward(self, x):
        return self.classifier(self.feature(x))

model = LeNet5V1()
```

```python
summary(model=model, 
        input_size=(1, 1, 28, 28), 
        col_width=20,
        col_names=['input_size', 'output_size', 'num_params', 'trainable'], 
        row_settings=['var_names'], 
        verbose=0)
```

```python
model = LeNet5V1()
x = torch.randn(1, 1, 28, 28)
y = model(x)
graph = make_dot(y, params=dict(model.named_parameters()), show_attrs=True, show_saved=True)
display(graph)
```

```python
loss_fn = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(params=model.parameters(), lr=0.001)
accuracy = Accuracy(task='multiclass', num_classes=10)
```

```python
from tqdm import tqdm
from torch.utils.tensorboard import SummaryWriter
from datetime import datetime
import os

timestamp = datetime.now().strftime("%Y-%m-%d")
experiment_name = "MNIST"
model_name = "LeNet5V1"
log_dir = os.path.join("runs", timestamp, experiment_name, model_name)
writer = SummaryWriter(log_dir)

# device-agnostic setup
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)  # Use the original variable name

EPOCHS = 12
train_loss_results = []
val_loss_results = []
train_acc_results = []
val_acc_results = []

for epoch in tqdm(range(EPOCHS)):
    train_loss, train_acc = 0.0, 0.0
    model.train()
    
    for X, y in train_dataloader:
        X, y = X.to(device), y.to(device)
        
        y_pred = model(X)
        loss = loss_fn(y_pred, y)
        train_loss += loss.item()
        
        acc = accuracy(y_pred, y)
        train_acc += acc
        
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
    train_loss /= len(train_dataloader)
    train_acc /= len(train_dataloader)
    train_loss_results.append(train_loss)
    train_acc_results.append(train_acc)
        
    val_loss, val_acc = 0.0, 0.0
    model.eval()
    with torch.inference_mode():
        for X, y in val_dataloader:
            X, y = X.to(device), y.to(device)
            
            y_pred = model(X)
            loss = loss_fn(y_pred, y)
            val_loss += loss.item()
            acc = accuracy(y_pred, y)
            val_acc += acc
            
        val_loss /= len(val_dataloader)
        val_acc /= len(val_dataloader)
        val_loss_results.append(val_loss)
        val_acc_results.append(val_acc)
        
    writer.add_scalars(main_tag="Loss", tag_scalar_dict={"train/loss": train_loss, "val/loss": val_loss}, global_step=epoch)
    writer.add_scalars(main_tag="Accuracy", tag_scalar_dict={"train/acc": train_acc, "val/acc": val_acc}, global_step=epoch)
    
    print(f"Epoch: {epoch}| Train loss: {train_loss: .5f}| Train acc: {train_acc: .5f}| Val loss: {val_loss: .5f}| Val acc: {val_acc: .5f}")
```

```python
from pathlib import Path

MODEL_PATH = Path("models")
MODEL_PATH.mkdir(parents=True, exist_ok=True)

MODEL_NAME = "lenet5_v1_mnist.pth"
MODEL_SAVE_PATH = MODEL_PATH / MODEL_NAME

# Saving the model
print(f"Saving the model: {MODEL_SAVE_PATH}")
torch.save(obj=model.state_dict(), f=MODEL_SAVE_PATH)

# Loading the saved model
model_lenet5_v1_mnist_loaded = LeNet5V1()
model_lenet5_v1_mnist_loaded.load_state_dict(torch.load(MODEL_SAVE_PATH))
```

```python
test_loss, test_acc = 0, 0

model_lenet5_v1_mnist_loaded.to(device)

model_lenet5_v1_mnist_loaded.eval()
with torch.inference_mode():
    for X, y in test_dataloader:
        X, y = X.to(device), y.to(device)
        y_pred = model_lenet5_v1_mnist_loaded(X)
        
        test_loss += loss_fn(y_pred, y)
        test_acc += accuracy(y_pred, y)
        
    test_loss /= len(test_dataloader)
    test_acc /= len(test_dataloader)

print(f"Test loss: {test_loss: .5f}| Test acc: {test_acc: .5f}")
```

```python
torch.manual_seed(42)  # setting random seed
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure(figsize=(12, 4))

rows, cols = 2, 6
for i in range(1, (rows * cols) + 1):
    random_idx = torch.randint(0, len(test_dataset), size=[1]).item()
    img, label_gt = test_dataset[random_idx]
    img_temp = img.unsqueeze(dim=0).to(device)
    # print(img.shape)
    label_pred = torch.argmax(model_lenet5_v1_mnist_loaded(img_temp))
    fig.add_subplot(rows, cols, i)
    img = img.permute(1, 2, 0)    # CWH --> WHC
    plt.imshow(img, cmap='gray')
    if label_pred == label_gt:
        plt.title(label_gt, color='g') # for correct prediction
    else:
        plt.title(label_gt, color='r') # for incorrect prediction
    plt.axis(False)
    plt.tight_layout();
```

### Step 2: Fully Connected Neural Network (FNN) Mode

#### Implement an FNN with 2 hidden layers, using ReLU activations, followed by a softmax output layer.

```python
#Reference: https://keras.io/guides/sequential_model/

# Create the FNN model
FNNmodel = keras.Sequential(
    [
    keras.Input(shape=(784,)), # Input layer (28 * 28 = 784)
    layers.Dense(128, activation='relu'), # First hidden layer with 128 units
    layers.Dense(64, activation='relu'), # Second hidden layer with 64 units
    layers.Dense(10, activation='softmax', name="output_layer") # Output layer with 10 units (one for each digit)
    ]
)
```

```python
# Print model summary
FNNmodel.summary()
```

```python
# Compile the model
FNNmodel.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model (for example, 10 epochs)
FNNmodel.fit(x_train, y_train, epochs=10)

# Call the model on a test input to see the output shape (for one image)
output = FNNmodel.predict(x_train[:1])

print(output)
```

#### Add impulsive noise to a selected number of test images (choose a number and specify it in your report), then evaluate the FNN model on the noisy test data and record the performance metrics.

```python
def add_impulsive_noise(image, noise_factor=0.1):
    """
    Add impulsive noise to an image.
    Parameters:
        - image: The image (2D or 3D array) to add noise to.
        - noise_factor: Proportion of image pixels to modify.
    """
    np.random.seed(42)

    noisy_image = image.copy()

    #set some pixels to 1
    white_pixels = np.random.rand(*image.shape) < (noise_factor / 2)
    noisy_image[white_pixels] = 1

    # set some pixels to 0
    black_pixels = np.random.rand(*image.shape) < (noise_factor / 2)

    noisy_image[black_pixels] = 0

    return noisy_image
```

```python
num_noisy = 1000
x_test_noisy = x_test.copy()

for i in range(num_noisy):
    x_test_noisy[i] = add_impulsive_noise(x_test[i], noise_factor=0.1)


image_shape = x_train.shape[1:]
flattened_size = np.prod(image_shape)
x_train_flat = x_train.reshape(-1, flattened_size)
x_test_flat = x_test.reshape(-1, flattened_size)
x_test_noisy_flat = x_test_noisy.reshape(-1, flattened_size)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input

model = Sequential([
    Input(shape=(28*28,)),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(x_train_flat, y_train, epochs=40, batch_size=32, validation_split=0.1)
```

```python
loss, accuracy = model.evaluate(x_test_noisy_flat, y_test, verbose=0)
print(f"Noisy Test Accuracy: {accuracy:.4f}")
```

```python
import shap
import numpy as np
import matplotlib.pyplot as plt

def explain_clean_predictions(model, x_test_flat, num_samples=10):
    # Select test samples
    test_samples = x_test_flat[:num_samples]  # Shape: (10, 784)
    
    # Compute SHAP values
    explainer = shap.DeepExplainer(model, background)  # background: e.g., (100, 784)
    shap_values = explainer.shap_values(test_samples)  # List of 10 arrays, each (10, 784)
    
    # Verify shapes
    print(f"test_samples shape: {test_samples.shape}")
    print(f"Length of shap_values: {len(shap_values)}")
    for k, sv in enumerate(shap_values):
        print(f"shap_values[{k}].shape: {sv.shape}")
    
    # Summary plot for all classes
    shap.summary_plot(
        shap_values,
        test_samples,
        feature_names=[f"pixel_{i}" for i in range(test_samples.shape[1])],
        class_names=[str(i) for i in range(10)],
        show=False
    )
    plt.title("SHAP Summary Plot - Clean Data (All Classes)")
    plt.show()
    
    # Example: Force plot for the first sample’s predicted class
    predictions = model.predict(test_samples)  # Shape: (10, 10)
    predicted_class = np.argmax(predictions[0])  # First sample’s predicted class
    shap.force_plot(
        explainer.expected_value[predicted_class],
        shap_values[predicted_class][0],  # SHAP values for predicted class, sample 0
        test_samples[0],
        matplotlib=True,
        show=False
    )
    plt.title(f"SHAP Force Plot - Clean Data (Sample 0, Class {predicted_class})")
    plt.show()
    
    return shap_values

# Assuming background and model are defined
background = x_train_flat[np.random.choice(x_train_flat.shape[0], 100, replace=False)]
clean_shap_values = explain_clean_predictions(model, x_test_flat)
```
